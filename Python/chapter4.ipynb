```python
#lists and tuples
friends = ["apple","orange",5,345.06,"mohit","xyz"]
#in lists we can store any type of datatype
print(friends[0])
#we can change in lists
#lists are mutable
friends[0]= "graphes"

print(friends[0])
```

```python
# Creating a list
fruits = ["apple", "banana", "cherry"]

# Accessing elements
print(fruits[0])   # apple
print(fruits[1])   # banana

# Modifying list
fruits[1] = "mango"
print(fruits)      # ['apple', 'mango', 'cherry']

# Adding element
fruits.append("orange")
print(fruits)      # ['apple', 'mango', 'cherry', 'orange']

# Removing element
fruits.remove("apple")
print(fruits)      # ['mango', 'cherry', 'orange']
```

```python
# Creating a tuple
#uple Example (Immutable)

#Tuples cannot be changed once created.
numbers = (10, 20, 30, 40)

# Accessing elements
print(numbers[0])   # 10
print(numbers[2])   # 30

# Tuples cannot be modified (this will cause error)
# numbers[1] = 50   ‚ùå Not allowed

# But you can do this:
new_tuple = numbers + (50, 60)  
print(new_tuple)    # (10, 20, 30, 40, 50, 60)

```

```python
# Creating a tuple
my_tuple = (10, 20, 30)

# Trying to change the first element
try:
    my_tuple[0] = 100
except TypeError as e:
    print("Error:", e)
```
