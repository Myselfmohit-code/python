```python
# Open file in write mode (creates new file or overwrites if exists)
f = open("example.txt", "w")
f.write("Hello, this is my first file handling program!\n")
f.write("Python makes it easy.")
f.close()
```

```python
# Open file in read mode
f = open("example.txt", "r")
content = f.read()   # reads whole file
print(content)
f.close()
```

```python
f = open("example.txt", "r")
for line in f:
    print(line.strip())   # strip removes extra spaces/newlines
f.close()
```

```python
f = open("example.txt", "a")  # 'a' means append
f.write("\nThis line is added later.")
f.close()
```

```python
with open("example.txt", "r") as f:
    print(f.read())
```

```python
def add_student():
    with open("students.txt", "a") as f:
        name = input("Enter student name: ")
        roll = input("Enter roll number: ")
        f.write(f"{roll}, {name}\n")
    print("Student added successfully!")

def view_students():
    with open("students.txt", "r") as f:
        print("\n--- Student Records ---")
        print(f.read())

# Example Usage
add_student()
view_students()
```

```python
with open("data.txt", "r") as f:
    lines = f.readlines()

unique_lines = list(set(lines))   # remove duplicates

with open("data.txt", "w") as f:
    f.writelines(unique_lines)
```

```python
word = input("Enter word to search: ")

with open("data.txt", "r") as f:
    data = f.read()
    if word in data:
        print("Word found!")
    else:
        print("Word not found.")
```
