```python
i=1

while(i<6):
    print(i)
    i+=1
```

```python
 #lists using while loops

l=[1,"mohit",False,"this","rohan","subham"]

i=0

while(i<len(l)):
    print(l[i])
    i +=1
```

```python
#for loops with else

l=[1,2,3]

for item in l:
    print(item)

else:
    print("done")
    
```

```python
#break

for i in range (100):
    if(i==34):
        break #exit loops right now
    print(i)


    
```

```python

#continue
for i in range (100):
    
    if(i==34):
        continue #skip iteration 
    print(i)
```
